# OpenAI 库基础使用

## 获取客户端对象

- 方法：`OpenAI()`
- 参数：
  - `api_key`: 密钥，从环境变量 `DASHSCOPE_API_KEY` 中获取
  - `base_url`: 请求地址
  - `timeout`: 请求超时时间
  - `max_retries`: 最大重试次数

```python
import os
from openai import OpenAI

client= OpenAI = OpenAI(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)
```

## 调用模型

- 方法：`client.chat.completions.create(...)`
- 参数：
  - `model`: 模型名称
  - `messages`: 聊天内容
  - `stream`: 是否流式返回，默认 false
  - `temperature`: 采样温度，范围 [0, 2)
  - `max_tokens`: 最大生成长度
  - `stop`: 停止符
- `message` 角色说明
  - `system`: 系统角色，设定整体行为和规则
  - `assistant`:代表 AI 的回答
  - `user`: 用户角色

```python
response = client.chat.completions.create(
   model="qwen3-max",
   messages=[
       {"role": "system", "content": "你是一个 Python 编程专家，并且不说废话简单回答"},
       {"role": "assistant", "content": "好的，我是编程专家，并且话不多，你要问什么？"},
       {"role": "user", "content": "输出 1-10 的数字，使用 python 代码"}
   ]
)
```

## 获取结果

- 属性：
  - `id`: 请求 ID
  - `object`: 对象类型
  - `created`: 创建时间
  - `model`: 模型名称
  - `usage`: 使用情况
  - `choices`: 模型返回结果

```python
from IPython.display import JSON, display

display(JSON(response.to_dict()))
```

## 获取结果内容

- 方法：`response.choices[0].message.content`
- 属性：
  - `index`: 索引
  - `message`: 模型返回结果
    - `role`: 角色
    - `content`: 内容
  - `finish_reason`: 结束原因

```python
print(response.choices[0].message.content)
```
