# LangChain 访问 Qwen 模型

- LangChain 支持的模型类型
  - LLMs：大语言模型
  - 聊天模型：转为场景对话优化的 LLMs
  - 文本嵌入模型：将文本转换为向量

```python
from langchain_community.llms.tongyi import Tongyi
```

```python
# qwen3-max 是聊天模型，qwen-max 是 LLM
llm = Tongyi(model="qwen-max")

res = llm.invoke("帮我讲个笑话")
```
